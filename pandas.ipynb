# Handling Missing Data (Dropena and fillna) | Pandas tutorial

 # Dropena function 

```python
import pandas as pd
d=pd.read_csv("W:\\islamabad\excel file\\missing_data_my_csv_file.csv")
d.head(10)


```

```python
d.isnull().sum()  # Check for missing values
```

```python
d1=d.dropna() 
d1.head(5)  # all missing data khtam
```

```python
d2=d.dropna(axis=0)    # 0 represent row and 1 represent column
d2.head(5)     # specific row or columns khtam the missing data then we used 
```

```python
d3=d.dropna(axis=1)    # 1 represent column
d3.head(5) 
```

```python
d3=d.dropna(how='all')    # this will only remove the null all row not if one null in a row
d3.head(5)
```

```python
d4=d.dropna(subset='Profit')   # remove a specific columns null value
d4.head(5) 
```

# fillna

```python
d5=d.fillna('walid khan')
d5.head(5)
```

```python
d6=d.fillna({'Ship Mode':'khan', 'Customer Segment':'jan'})  # if each column we have to give values with null
d6.head(5)
```

## forward fill the column mean che km ke null ye hga ba lapasa values rashy

```python
d7=d.fillna(method='ffill',axis=0)  # if each column we have to give values with null
d7.head(5)
```

### forward fill the column mean che km ke null ye hga ba lapasa values rashy

```python
d8=d.fillna(method='bfill',axis=1)  # if each column we have to give values with null
d8.head(5)
```

## Replace Function ( if you to replce any values in data set then we usd replace function

```python
import pandas as pd
s=pd.read_csv("W:\\excel file\\missing_data_my_csv_file.csv")
s.head(5)
```

## Groupby   : used kaw che kala gad wada data ye aw te similer thing group julaw

```python
import pandas as pd
var={
    'name':['a','b','c','a','c','b','b','c','a','b'],  # soppose ye subject ke name hai same data hai
    'sub1':[10,20,30,23,34,45,56,67,67,78],   # two subject ke marks hai
    'sub2':[145,10,90,73,84,55,26,64,67,78]

}

var1=pd.DataFrame(var)
var1
```

similer ke group banaw ma

```python
var2=var1.groupby('name')
var2
```

```python
for sub1,sub2 in var2:
 print(sub1)
 print(sub2)
 print()
```

```python
# specific one group i want
var3=var2.get_group('a')
var3
```

```python
import pandas as pd

# Sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)

# Display the DataFrame
print("Original DataFrame:")
print(df)

# Select the first row using iloc
first_row = df.iloc[0]
print("\nFirst row using iloc:")
print(first_row)
```

```python
import pandas as pd

# Sample DataFrame with labeled index
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data, index=['a', 'b', 'c'])

# Display the DataFrame
print("Original DataFrame:")
print(df)

# Select the row with label 'a' using loc
row_a = df.loc['c']
print("\nRow with label 'a' using loc:")
print(row_a)
```


# Use iloc when you need to select data based on the numerical position.
# Use loc when you need to select data based on labels or boolean conditions.
